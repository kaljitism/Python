```python
from time import sleep
```

```python
class Hello(Thread):
    def run(self):
        for i in range(500): 
            print("Hello")
            sleep(1)
            
class Hi(Thread):
    def run(self):
        for i in range(500): 
            print("Hi")
            sleep(1)
```

```python
t1 = Hello()
t2 = Hi()
```

```python
t1.start()
sleep(0.2)
t2.start()
```
